```python
import numpy as np
numeric=df.select_dtypes(include=np.number)
print(numeric)
numeric_columns=numeric.columns
print(numeric_columns)
```

```python
df[numeric_columns]=df[numeric_columns].fillna(df[numeric_columns].median())
```

```python
df.isna().sum()
```

```python
boolean_columns=df.select_dtypes(include=object).columns.tolist()
print(boolean_columns)
boolean_columns.remove("class")
print(boolean_columns)
df[boolean_columns]=df[boolean_columns].astype("bool")
```

```python
df[boolean_columns]=df[boolean_columns].fillna(df[boolean_columns].mode())
```

```python
df["Tweet Location"]=df["Tweet Location"].astype("string")
```

```python
df.dtypes
```

```python
import re  #regular expressions manipulates strings 
def set_pattern(x):
    pattern=r'[(A-Z)]\w+,([A-Z])\w+' #w+ means many letter preceded by a capital letter
    res=re.match(pattern, x)
    if res:
        x=x.replace(",", ", ")
    return x
```

```python
df["Tweet Location"]=df.apply(lambda x: set_pattern(x["Tweet Location"]), axis=1) #much faster
```
